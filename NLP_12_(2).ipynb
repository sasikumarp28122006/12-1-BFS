```python
graph = {

'A': ['B','C'],

'B': ['D','E'],

'C': ['F'],

'D': [],

'E': ['F'],

'F': []

}

visited = []
```

```python
visited = []
def dfs (node):
    if node not in visited:
        print(node, end=" ")
        visited.append(node)
    for neighbour in graph [node]:
        dfs (neighbour)

print("DFS Traversal:")
dfs('A')
```
